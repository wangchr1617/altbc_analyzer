### 调用 ALTBC_Analyzer 分析轨迹文件

```python
from altbc_analyzer.altbc_analyzer import ALTBC_Analyzer
import matplotlib.pyplot as plt
filename = './example/test.xyz'
analyzer = ALTBC_Analyzer(filename, filetype='MultiSystem', frame_interval=1, grid_size=0.005, center_element='Ge')
plt.figure(figsize=(6, 5))
analyzer.plot_grid()
plt.savefig('./altbc1.png', bbox_inches='tight')
```

### 调用 ALTBC_Analyzer 分析某一帧轨迹文件

```python
from altbc_analyzer.altbc_analyzer import ALTBC_Analyzer
filename = './example/POSCAR'
analyzer = ALTBC_Analyzer(filename, filetype='SingleSystem', center_element='Ge')
plt.figure(figsize=(6, 5))
analyzer.plot_scatter()
plt.savefig('./altbc2.png', bbox_inches='tight')
```
